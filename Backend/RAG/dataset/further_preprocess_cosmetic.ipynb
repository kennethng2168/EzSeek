```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('./dataset/cosmetic_clean_final.csv')

# Display the dataframe
print(df)
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each category
category_counts = df['category'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
category_counts.plot.pie(autopct='%1.1f%%')
plt.title('Category Distribution')
plt.ylabel('')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each subcategory
subcategory_counts = df['subcategory'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 6))
subcategory_counts.plot.pie(autopct='%1.1f%%')
plt.title('Subcategory Distribution')
plt.ylabel('')
plt.show()
```
